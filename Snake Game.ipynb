```python
import random

# Set up the game board
board_width = 20
board_height = 10

# Snake initial position and size
snake_x = board_width // 2
snake_y = board_height // 2
snake_length = 1
snake_body = [(snake_x, snake_y)]

# Food position
food_x = random.randint(0, board_width - 1)
food_y = random.randint(0, board_height - 1)

# Game loop
game_over = False
while not game_over:
    # Display the game board
    for y in range(board_height):
        for x in range(board_width):
            if (x, y) == (food_x, food_y):
                print('F', end=' ')
            elif (x, y) == snake_body[0]:
                print('H', end=' ')
            elif (x, y) in snake_body[1:]:
                print('S', end=' ')
            else:
                print('.', end=' ')
        print()

    # Get user input for snake direction
    direction = input("Enter direction (up/down/left/right): ")

    # Update snake position
    if direction == 'up':
        snake_y -= 1
    elif direction == 'down':
        snake_y += 1
    elif direction == 'left':
        snake_x -= 1
    elif direction == 'right':
        snake_x += 1

    # Check for collision with food
    if (snake_x, snake_y) == (food_x, food_y):
        snake_length += 1
        food_x = random.randint(0, board_width - 1)
        food_y = random.randint(0, board_height - 1)

    # Update snake body
    snake_body.insert(0, (snake_x, snake_y))
    if len(snake_body) > snake_length:
        snake_body.pop()

    # Check for collision with walls or snake's own body
    if snake_x < 0 or snake_x >= board_width or snake_y < 0 or snake_y >= board_height or (snake_x, snake_y) in snake_body[1:]:
        game_over = True

print("Game Over")
```
